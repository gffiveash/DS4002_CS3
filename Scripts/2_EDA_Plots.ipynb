```python
# Import necessary packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading file
df=pd.read_csv('DS4002_CS3/Data/combined_clean_data.csv')
```

##Violin Plots

```python
# Create violin plot, illustrating the sentiment distribution for celebrity vs. non-celebrity makeup brands
plt.figure(figsize=(10, 6))
sns.violinplot(x='CelebrityBranded', y='CompoundScore', data=df, palette='Set1')
plt.title('Sentiment Distribution of Celebrity vs Non-Celebrity Makeup Brands')
plt.xlabel('Celebrity Branded?')
plt.ylabel('Compound Sentiment Score')
plt.ylim(-1, 1)

plt.show()
```

```python
# Create violin plot, illustrating the sentiment distribution across each makeup brand
plt.figure(figsize=(12, 6))
sns.violinplot(x='Brand', y='CompoundScore', hue='CelebrityBranded', data=df, split=True, inner='quartile', palette='Set2')
plt.title('Sentiment Distribution Across Makeup Brands')
plt.xlabel('Brand', fontsize=12)
plt.ylabel('Compound Sentiment Score')
plt.ylim(-1, 1)

plt.show()
```

##Boxplot

```python
# Creating a boxplot that illustrates the distribution of sentiment for celebrity vs. non-celebrity brands
plt.figure(figsize=(8, 6))
sns.boxplot(x='CelebrityBranded', y='CompoundScore', data=df, hue='CelebrityBranded')
plt.title('Sentiment Polarization for Celebrity vs Non-Celebrity Brands')
plt.show()
```

##Bar Graphs

```python
# Creating a bar graph capturing the distribution of VADER sentiment analysis scores
plt.figure(figsize=(10, 6))
sns.histplot(df, x="CompoundScore", hue="Brand", bins=30, kde=True, alpha=0.5)
plt.title("Distribution of Compand Sentiment Scores")
plt.xlabel("VADER Compound Sentiment Score")
plt.ylabel("Frequency")
plt.legend(title="Brand Type")
plt.show()
```

```python
# Bar graph of sentiment (positive, neutral, negative) hued by celebrity branded vs. non-celebrity associated
df["sentiment_category"] = pd.cut(df["CompoundScore"],bins=[-1, -0.05, 0.05, 1], labels=["Negative", "Neutral", "Positive"])
df["CelebrityBranded"] = df["CelebrityBranded"].astype(str)
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x="sentiment_category", hue="CelebrityBranded")
plt.title("Sentiment (Positive, Neutral, Negative) by Brand Type")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.legend(title="Celebrity Branded")
plt.show()
```
